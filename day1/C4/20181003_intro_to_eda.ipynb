# Distribution of the Data



```python
df.SalePrice.describe()
```

```python
plt.figure(figsize=(15,10))
sns.distplot(df.SalePrice)
plt.title("Distribution of SalePrice", fontsize = 15)
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(df.SalePrice)
plt.title("Distribution of SalePrice", fontsize = 15)
```

```python
# Two subplots, unpack the axes array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize = (10,6))
ax1.boxplot(df.SalePrice, whis=1.5)
ax1.set_title("SalePrice")
ax1.grid()
ax2.violinplot(dataset= df.SalePrice,
               showmedians=True)
ax2.grid()
plt.show()
```

```python
# add transparency
df.plot(kind='scatter', x='LotArea', y='SalePrice', alpha=0.3, figsize = (15,10))
plt.title("LotArea vs SalePrice")
```

```python

# add transparency
df.plot(kind='scatter', x='MSSubClass', y='SalePrice', alpha=0.3, figsize = (15,10))
plt.title("MSSubClass vs SalePrice")
```

```python
# scatter matrix of three numerical columns

pd.scatter_matrix(df[['SalePrice', 'MSSubClass', 'LotArea']], figsize=(15,15))
plt.show()
```

# Look for Corellation among the features with Heatmap

```python
f, ax = plt.subplots(figsize=(20,20))
sns.heatmap(df.corr(), annot=True, linewidths=0.5, fmt='.1f', ax=ax, )
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```
